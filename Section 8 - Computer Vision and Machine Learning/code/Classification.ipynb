# Feature Extraction

Create a feature that represents the brightness in an image. We'll be extracting the **average brightness** using HSV colorspace. Specifically, we'll use the V channel (a measure of brightness), add up the pixel values in the V channel, then divide that sum by the area of the image to get the average Value of the image.


---
###  Find the average brightness using the V channel

This function takes in a **standardized** RGB image and returns a feature (a single value) that represent the average level of brightness in the image. We'll use this value to classify the image as day or night.

````python
# Find the average Value or brightness of an image
def avg_brightness(rgb_image):
    # Convert image to HSV
    hsv = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)

    # Add up all the pixel values in the V channel
    sum_brightness = np.sum(hsv[:,:,2])
    area = 600*1100.0  # pixels
    
    # find the avg
    avg = sum_brightness/area
    
    return avg
````

````python
def get_label_list():
    train_label_list = []
    for i in range(len(IMAGE_LIST)):
        if IMAGE_LIST[i][1] == 'day':
            train_label_list.append(1)
        else:
            train_label_list.append(0)
    return train_label_list

training_labels = get_label_list()
training_labels[0:5]
````

````python
def accuracy_score(y_true, y_pred):
    counter = 0
    for index, label in enumerate(y_true):
        if label == y_pred[index]:
            counter += 1
    print(counter)
    print(len(y_true))
    return counter/len(y_true)
````

````python
def train():
    accuracy = 0
    for brightness_param in range(20, 70):
        train_predictions = []
        for image in IMAGE_LIST:
            if avg_brightness(image[0]) < brightness_param:
                train_predictions.append(0)
            else:
                train_predictions.append(1)
        new_accuracy = accuracy_score(training_labels, train_predictions)
        if accuracy < new_accuracy:
            accuracy = new_accuracy
    return accuracy, brightness_param
````

````python
def predict(brightness_param):
    train_predictions = []
    for image in IMAGE_LIST:
        if avg_brightness(image[0]) < brightness_param:
            train_predictions.append(0)
        else:
            train_predictions.append(1)
    return train_predictions
````

````python
accuracy, brightness_value = train()
print(accuracy)
````
